```python
#ä½¿ç”¨æ¯å°æ—¶çš„æ¸©åº¦æ•°æ®
hourly_temp = hourly['T (degC)']
len(hourly_temp)
```

```python
#To predict what the temperature will be in 24 hours, I use the offset parameter

#[[[1],[2],[3],[4],[5]]][5+24]
#[[[2],[3],[4],[5],[6]]][6+24]
#[[[3],[4],[5],[6],[7]]][7+24]
#ä½¿ç”¨1è‡³5æ—¶çš„æ•°æ®é¢„æµ‹29æ—¶çš„æ•°æ®ï¼Œ2è‡³6æ—¶çš„æ•°æ®é¢„æµ‹30æ—¶çš„æ•°æ®
def generator(data, window, offset):
    gen = data.to_numpy()
    X = []
    y = []
    for i in range(len(gen)-window-offset):
        row = [[a] for a in gen[i:i+window]]
        X.append(row)
        label = gen[i+window+offset-1]
        y.append(label)
    return np.array(X), np.array(y)
```

```python
WINDOW = 5
OFFSET = 24
X, y = generator(hourly_temp, WINDOW, OFFSET)
X.shape, y.shape
```

```python
#Train, valid and test sets
#å°†æ•°æ®é›†åˆ†ä¸ºè®­ç»ƒé›†ã€éªŒè¯é›†å’Œæµ‹è¯•é›†
X_train, y_train = X[:60000], y[:60000]
X_val, y_val = X[60000:65000], y[60000:65000]
X_test, y_test = X[65000:], y[65000:]
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
#ä½¿ç”¨Kerasè°ƒç”¨TensorFlow

from keras.models import Sequential
from keras.layers import *
from keras.callbacks import ModelCheckpoint
from keras.losses import MeanSquaredError
from keras.metrics import RootMeanSquaredError
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
```

```python
#ä½¿ç”¨ç®€å•çš„é¡ºåºæ¨¡åž‹
#ç¬¬ä¸€å±‚æ˜¯è¾“å…¥å±‚
#ç¬¬äºŒå±‚æ˜¯æœ‰ 100 ä¸ªç¥žç»å…ƒå’Œæ¿€æ´»å‡½æ•°çš„ LSTM å±‚
#ç¬¬ä¸‰å±‚æ˜¯8ä¸ªè¾“å‡ºçš„å…¨è¿žæŽ¥å±‚ï¼Œæ¿€æ´»å‡½æ•°æ˜¯relu
#ç¬¬å››å±‚æ˜¯1ä¸ªè¾“å‡ºçš„å…¨è¿žæŽ¥å±‚ï¼Œä½¿ç”¨çº¿æ€§æ¿€æ´»
#LSTM
model1 = Sequential()
model1.add(InputLayer((WINDOW, 1)))
model1.add(LSTM(100))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python

cp = ModelCheckpoint('model1/', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, callbacks=[cp])
```

```python
from keras.models import load_model

model1 = load_model('model1/')
```

```python
test_predictions = model1.predict(X_test).flatten()

result = pd.DataFrame(data={'Predicted': test_predictions, 'Real':y_test})
plt.figure(figsize=(20,7.5),dpi=120)
plt.plot(result['Predicted'][:300], "-g", label="Predicted")
plt.plot(result['Real'][:300], "-r", label="Real")
plt.legend(loc='best')
```
